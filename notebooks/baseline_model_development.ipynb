#### Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Load processed datasets

```python
training_set_path = "../data/processed/training_boston.csv"
test_set_path = "../data/processed/test_boston.csv"

training_set = pd.read_csv(training_set_path)
test_set = pd.read_csv(test_set_path)

X_train = training_set.drop(columns = ["MEDV"])
X_test = test_set.drop(columns = ["MEDV"])
y_train = training_set["MEDV"]
y_test = test_set["MEDV"]
```

#### Defining baseline linear regression model

```python
from sklearn.linear_model import LinearRegression

regression_model = LinearRegression()
regression_model.fit(X_train, y_train)

y_pred = regression_model.predict(X_test)
```

#### Evaluating the prediction performance metrics
- Our main goal is to come up with a more complex regression model where we outperform our baseline linear regression model

```python
from sklearn.metrics import root_mean_squared_error, r2_score

regression_rmse = root_mean_squared_error(y_test, y_pred)
regression_r2 = r2_score(y_test, y_pred)

print(f"Baseline Linear Regression Model RMSE: {regression_rmse}")
print(f"Baseline Linear Regression Model R²: {regression_r2}")
```

#### Visualize actual vs pred

```python
# visualize
plt.figure(figsize=(10, 6))
# alpha param to adjust transparency
plt.scatter(y_test, y_pred, alpha=0.7)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], '--', color='red')
plt.xlabel("Actual")
plt.ylabel("Pred")
plt.title("Actual vs Predicted")
plt.show()
```
